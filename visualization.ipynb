```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset('iris')
df
```

```python
sns.barplot()
```

```python
sns.barplot(df, x='species', y= 'sepal_length')
```

```python
sns.barplot(df, x='species', y= 'sepal_width')
```

```python
fig, ax = plt.subplots(figsize=(4,4))
sns.barplot(df, x='species', y= 'sepal_width', ax=ax)
```

```python
fig, ax = plt.subplots(figsize=(20,4), ncols=4)
```

```python
fig, ax = plt.subplots(figsize=(20,4), nrows=4)
```

```python
fig, ax = plt.subplots(figsize=(20,4), ncols=4)
sns.barplot(df, x='species',y='sepal_length', ax=ax[0])
sns.barplot(df, x='species',y='sepal_length', ax=ax[1])
sns.barplot(df, x='species',y='sepal_length', ax=ax[2])
sns.barplot(df, x='species',y='sepal_length', ax=ax[3])
```

```python
fig, ax = plt.subplots(figsize=(8,8), ncols=2, nrows=2)
sns.barplot(df, x='species',y='sepal_length', ax=ax[0,0])
sns.barplot(df, x='species',y='sepal_length', ax=ax[0,1])
sns.barplot(df, x='species',y='sepal_length', ax=ax[1,0])
sns.barplot(df, x='species',y='sepal_length', ax=ax[1,1])
```

```python
sns.scatterplot(df, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.jointplot(df, x='sepal_length', y='sepal_width', hue='species')
```

```python
sns.jointplot(df, kind='hex',x= 'sepal_length', y='sepal_width')
```
