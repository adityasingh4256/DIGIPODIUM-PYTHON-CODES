```python
def clean_price_column(price):
    if isinstance(price, str):
        price = price.replace(',','')
        price = float(price)
        return price
    return price
```

```python
# test

clean_price_column('12,43,4555,333')
```

```python
# testing the function on the column

df['price'].apply(clean_price_column)
```

```python
# implementation

df['clean_price'] = df['price'].apply(clean_price_column)
df
```

```python
def clean_rating_column(rating):
    if isinstance(rating, str):
        if len(rating) <=3:
            rating = float(rating)
            return rating
        else:
            return np.nan
        return rating
```

```python
# test
clean_rating_column("no review")
```

```python
#testing
df['rating'].apply(clean_rating_column)
```

```python
df['clean_rating'] = df['rating'].apply(clean_rating_column)
df
```

```python
def clean_oprice_column(oprice):
    if isinstance(oprice, str):
        if len(oprice) <=3:
            oprice = float(oprice)
            return oprice
        else:
            return np.nan
        return oprice
```

```python
clean_oprice_column("oprice")
```

```python
def clean_oprice_column(oprice):
    if isinstance(oprice, str):
        oprice = oprice.replace(',','').replace('₹','')
        oprice = float(oprice)
        return oprice
    return oprice
```

```python
df['clean_oprice'] = df['oprice'].apply(clean_oprice_column)
df
```
